### Generate 1000 numbers according to given series equation: 
### $$n[k] = fractional\hspace{3pt} part\hspace{3pt} of\hspace{3pt} (n[k-1] + \pi ) \times 100$$

```python
import math
array = []
result = []
array.append(0.2)
result.append(0.2000)
```

The following loop calculates the elements and appends it into a result list

```python
for i in range(1,1000):
    array.append(((array[i-1] + math.pi)*100) - (int)((array[i-1] + math.pi)*100))
    result.append("%.4f" %array[i])
```

The result list is printed

```python
for i in range(0,len(result)):
    print ('%d: %s' %(i+1, result[i]))
```
