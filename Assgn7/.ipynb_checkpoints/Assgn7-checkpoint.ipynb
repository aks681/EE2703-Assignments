```python
import numpy as np
import scipy.signal as sp
import matplotlib.pyplot as plt
```

```python
num = np.poly1d([1, 0.5])
den = np.polymul([1, 1, 2.5],[1,0,2.25])
H = sp.lti(num,den)
t,x = sp.impulse(H,None,T=np.linspace(0,50,1001))
plt.plot (t,x)
plt.show()
```

```python
num = np.poly1d([1, 0.5])
den = np.polymul([1, 1, 2.5],[1,0,2.25])
H = sp.lti(num,den)
t = np.linspace(0.0,50.0,1001)
f = np.cos(1.5*t)*np.exp(-0.5*t)
t,y,svec=sp.lsim(H,f,t)
plt.plot(t,y)
plt.show()
```

```python
num = np.poly1d([1, 0.05])
den = np.polymul([1, 0.1, 2.2525],[1,0,2.25])
H = sp.lti(num,den)
t,x = sp.impulse(H,None,T=np.linspace(0,50,1001))
plt.plot(t,x)
plt.show()
```

```python
for w in np.linspace(1.4,1.6,5):
    num = np.poly1d([1, 0.05])
    den = np.polymul([1, 0.1, 0.0025 + w*w],[1,0,w*w])
    H = sp.lti(num,den)
    t = np.linspace(0.0,50.0,1001)
    f = np.cos(w*t)*np.exp(-0.05*t)
    t,y,svec=sp.lsim(H,f,t)
    plt.plot(t,y)
    plt.show()
```

```python
num = np.poly1d([1, 1, 2, 2])
den = np.poly1d([1, 0, 3, 0, 0])
X = sp.lti(num,den)
t,x = sp.impulse(X,None,T=np.linspace(0,20,1001))
plt.plot(t,x)
plt.show()
```

```python
num = np.polymul([1,1,2,2],[1,0,1])
num = np.polyadd(num,np.polymul([-1, -1],[1,0,3,0,0]))
den = np.poly1d([1,0,3,0,0])
Y = sp.lti(num,den)
t,y = sp.impulse(Y,None,T=np.linspace(0,20,1001))
plt.plot(t,y)
plt.show()
```
